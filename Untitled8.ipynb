##Section 2. 4

####1st requirement to include the country and city name in one data frame in Section 2

```python
# Selecting the required columns from the code mapping sheet and joining on the country code and city code
query1 = """
SELECT td.*, cd.City AS CityName, cd.Country AS CountryName
FROM telco_data AS td
LEFT JOIN code_data AS cd
ON td.CountryCode = cd.CountryCode AND td.CityCode = cd.CityCode
"""
```

```python
# Execute the query
merged_data = ps.sqldf(query1, locals())

# Display the updated DataFrame
print(merged_data.head())
print('Shape:',merged_data.shape)
```

#### 2nd requirementin Section 2: Standardizing the data in the column

```python
#naming the standardized gender column with GenderStand
query2 = """
SELECT *,
    CASE
        WHEN Gender IN ('male', 'm', 'Male') THEN 'Male'
        WHEN Gender IN ('female', 'f', 'Female') THEN 'Female'
        ELSE NULL
    END AS GenderStand
FROM merged_data
"""

# Execute the query
merged_data = ps.sqldf(query2, locals())

# Display the updated DataFrame
print(merged_data.head())
```

#### 3rd requirement in Section 2 of splitting the column named Phone Service and Information into two columns

```python
#The columns Phone Service and Information was split into two columns named Registration_Status and Phone_Number
#The INSTR was used to search in the Phone Service and Information column for the values Yes and No then it will return the corresponding status
#If Yes exists in the column Phone Service and Information then it extracts the phone number that follows the comma and removes any extra spaces using trim.
#INSTR("Phone Service and Information", ',') + 1 finds the comman in the substring and calculates the position to start with in the substring which is after the ','
#SUBSTR extarcts everything from that position and trim to remove any leading or trailing whitespace from the extracted substring
query3 = """
SELECT
    *,
    CASE
        WHEN INSTR("Phone Service and Information", 'Yes') > 0 THEN 'Yes'
        WHEN INSTR("Phone Service and Information", 'No') > 0 THEN 'No'
        ELSE NULL
    END AS Registration_Status,
    CASE
        WHEN INSTR("Phone Service and Information", 'Yes') > 0 THEN TRIM(SUBSTR("Phone Service and Information", INSTR("Phone Service and Information", ',') + 1))
        ELSE NULL
    END AS Phone_Number
FROM merged_data
"""

# Execute the query
merged_data = ps.sqldf(query3, locals())

# Display the updated DataFrame
print(merged_data.head())
```

##Section 2. 5
Write a query to calculate the average monthly charge for customers across each country.

```python
#Calculating the average monthly charges for each country and grouping by the country
query3 = """
SELECT
    CountryName,
    AVG("Monthly Charges") AS Averga_monthly_charges
FROM merged_data
GROUP BY CountryName
"""
result3 = ps.sqldf(query3, locals())

print(result3)
```

##Section 2. 6
Write a query to find the top 5 customers with the highest and lowest CLTV.

```python
#Top 5 customers with lowest CLTV
query4 = """
SELECT
    CustomerID
FROM merged_data
Order BY CLTV
LIMIT 5
"""
result4 = ps.sqldf(query4, locals())

print(result4)
```

```python
#Top 5 customers with highest CLTV
query5 = """
SELECT
    CustomerID
FROM merged_data
Order BY CLTV DESC
LIMIT 5
"""
result5 = ps.sqldf(query5, locals())

print(result5)
```

##Section 2. 7
Write a query to find how many customers have a monthly charge above 6000.

```python
#the type of the Total CHarges was changed as it was an object
merged_data['Total Charges'] = pd.to_numeric(merged_data['Total Charges'], errors='coerce')
print(merged_data.dtypes)
```

```python
query5 = """
SELECT COUNT(DISTINCT CustomerID) AS Total_Customers
FROM merged_data
WHERE "Total Charges" > 6000
"""
result5 = ps.sqldf(query5, locals())
total_customers = result5['Total_Customers'].iloc[0]

print(f"The customers whose charges are above 6000 are: {total_customers} customers")
```
